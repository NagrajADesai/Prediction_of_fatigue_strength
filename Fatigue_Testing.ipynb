## Define dependent and independent variable

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X.shape
```

```python
y.shape
```

## Spliting data into train test and validation set

```python
from sklearn.model_selection import train_test_split
# for train and test
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1, random_state = 42)

# for validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.1, random_state = 42)
```

```python
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
print("X_val shape: {}".format(X_val.shape))
print("y_val shape: {}".format(y_val.shape))
```

## Training the Linear Regression model  

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
# coefficients and intercept
print(f" Coefficients\n {regressor.coef_}")
print(f"\n Intercept\n {regressor.intercept_}")
```

```python
# on which parameters the model has been trained
regressor.get_params()
```

## Accuracy on test set

```python
y_test_pred = regressor.predict(X_test)
```

```python
y_test_pred
```

```python
# R2 score
from sklearn.metrics import r2_score
print(r2_score(y_test, y_test_pred))
```

```python
# Mean Square Error(MSE) 
# Root Mean Square Error(RMSE)
# Mean Absolute Error(MAE)
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import math
print(f'Mean Square Error(MSE)- {mean_squared_error(y_test, y_test_pred)}')
print(f'Root Mean Square Error(RMSE)- {math.sqrt(mean_squared_error(y_test, y_test_pred))}')     
print(f'Mean Absolute Error(MAE)- {mean_absolute_error(y_test, y_test_pred)}')
```

## Accuracy on validation set

```python
y_val_pred = regressor.predict(X_val)
```

```python
# R2 score
print(r2_score(y_val, y_val_pred))
```

```python
# Mean Square Error(MSE) 
# Root Mean Square Error(RMSE)
# Mean Absolute Error(MAE)
print(f'Mean Square Error(MSE)- {mean_squared_error(y_val, y_val_pred)}')
print(f'Root Mean Square Error(RMSE)- {math.sqrt(mean_squared_error(y_val, y_val_pred))}')     
print(f'Mean Absolute Error(MAE)- {mean_absolute_error(y_val, y_val_pred)}')
```

## K-fold cross validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = regressor, X = X_train, y = y_train, cv = 10) #estimator = model
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
