## Define dependent and independent variable

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X.shape
```

```python
y.shape
```

## Spliting data into train test and validation set

```python
from sklearn.model_selection import train_test_split
# for train and test
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1, random_state = 42)

# for validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.1, random_state = 42)
```

```python
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
print("X_val shape: {}".format(X_val.shape))
print("y_val shape: {}".format(y_val.shape))
```

# Models

## Training the Linear Regression model  

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)

    # Accuracy on test set
y_test_pred = regressor.predict(X_test)

    # R2 score
print('Test set')
from sklearn.metrics import r2_score
print('R^2 score: {}'.format(r2_score(y_test, y_test_pred)))

    # Mean Square Error(MSE) 
    # Root Mean Square Error(RMSE)
    # Mean Absolute Error(MAE)
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import math
print(f'Mean Square Error(MSE)- {mean_squared_error(y_test, y_test_pred)}')
print(f'Root Mean Square Error(RMSE)- {math.sqrt(mean_squared_error(y_test, y_test_pred))}')     
print(f'Mean Absolute Error(MAE)- {mean_absolute_error(y_test, y_test_pred)}')

    # Accuracy on validation set
print('\nValidation set')
y_val_pred = regressor.predict(X_val)
    # R2 score
print('R^2 score: {}'.format(r2_score(y_val, y_val_pred)))

    # Mean Square Error(MSE) 
    # Root Mean Square Error(RMSE)
    # Mean Absolute Error(MAE)
print(f'Mean Square Error(MSE)- {mean_squared_error(y_val, y_val_pred)}')
print(f'Root Mean Square Error(RMSE)- {math.sqrt(mean_squared_error(y_val, y_val_pred))}')     
print(f'Mean Absolute Error(MAE)- {mean_absolute_error(y_val, y_val_pred)}')

    #K-fold cross validation
print("\nK-fold cross validation")
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = regressor, X = X_train, y = y_train, cv = 10) #estimator = model
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

## Decision tree

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(X_train,y_train)

ry_pred = dt.predict(X_test)

print('Test Set')
# R score
print('R^2 value: {}'.format(r2_score(y_test, ry_pred)))

# Mean Square Error(MSE) 
# Root Mean Square Error(RMSE)
# Mean Absolute Error(MAE)
print(f'Mean Square Error(MSE)- {mean_squared_error(y_test, ry_pred)}')
print(f'Root Mean Square Error(RMSE)- {math.sqrt(mean_squared_error(y_test, ry_pred))}')     
print(f'Mean Absolute Error(MAE)- {mean_absolute_error(y_test, ry_pred)}')

# validation set
print('\nValidation Set')
y_pred_val = dt.predict(X_val)

print("R^2 value:{}".format(dt.score(X_val, y_val)))
rmse = np.sqrt(mean_squared_error(y_val, y_pred_val))
print("rmse: {}".format(rmse))

# K-fold cross validation
print('\nK-fold cross validation Set')
accuracies = cross_val_score(estimator = dt, X = X_train, y = y_train, cv = 10) #estimator = model
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

## Random forest

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators= 100, random_state = 0)
rf.fit(X_train,y_train)

ry_pred = rf.predict(X_test)

print('Test Set')
# R score
print('R^2 value: {}'.format(r2_score(y_test, ry_pred)))

# Mean Square Error(MSE) 
# Root Mean Square Error(RMSE)
# Mean Absolute Error(MAE)
print(f'Mean Square Error(MSE)- {mean_squared_error(y_test, ry_pred)}')
print(f'Root Mean Square Error(RMSE)- {math.sqrt(mean_squared_error(y_test, ry_pred))}')     
print(f'Mean Absolute Error(MAE)- {mean_absolute_error(y_test, ry_pred)}')

# validation set
print('\nValidation Set')
y_pred_val = rf.predict(X_val)

print("R^2 value:{}".format(rf.score(X_val, y_val)))
rmse = np.sqrt(mean_squared_error(y_val, y_pred_val))
print("rmse: {}".format(rmse))

# K-fold cross validation
print('\nK-fold cross validation Set')
accuracies = cross_val_score(estimator = rf, X = X_train, y = y_train, cv = 10) #estimator = model
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

## Gradient Boost

```python
from sklearn.ensemble import GradientBoostingRegressor
gbm_model = GradientBoostingRegressor()
gbm_model.fit(X_train, y_train)
y_pred = gbm_model.predict(X_test)

print('Test Set')
print("R^2 value:{}".format(gbm_model.score(X_test, y_test)))
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("rmse: {}".format(rmse))


print('\nValidation Set')
y_pred_val = gbm_model.predict(X_val)


print("R^2 value:{}".format(gbm_model.score(X_val, y_val)))
rmse = np.sqrt(mean_squared_error(y_val, y_pred_val))
print("rmse: {}".format(rmse))

# K-fold cross validation
print('\nK-fold cross validation Set')
accuracies = cross_val_score(estimator = gbm_model, X = X_train, y = y_train, cv = 10) #estimator = model
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
